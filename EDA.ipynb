```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the visualization style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# Load the new dataset
df = pd.read_csv('world_bank_data_2025.csv')

# Display the first few rows and column names
print(df.columns.tolist())
df.head()
```

```python
# Check data types and missing values
print(df.info())

# Summary statistics for numerical columns
df.describe()
```

```python
# Select key numerical columns for correlation
cols_for_corr = [
    'Inflation (CPI %)', 'GDP Growth (% Annual)', 'Unemployment Rate (%)',
    'Interest Rate (Real, %)', 'Public Debt (% of GDP)', 'GDP per Capita (Current USD)',
    'Current Account Balance (% GDP)'
]

# Calculate correlation
corr_matrix = df[cols_for_corr].corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Key Economic Indicators', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
# Drop NA values for plotting
sns.histplot(df['GDP Growth (% Annual)'].dropna(), bins=40, kde=True, color='skyblue')
plt.title('Distribution of Annual GDP Growth (%)', fontsize=14)
plt.xlabel('GDP Growth (% Annual)')
plt.ylabel('Frequency')
plt.xlim(-15, 15) # Limiting x-axis to focus on the main distribution
plt.show()
```

```python
# Calculate average GDP per capita per year
avg_gdp_capita = df.groupby('year')['GDP per Capita (Current USD)'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(data=avg_gdp_capita, x='year', y='GDP per Capita (Current USD)', marker='o', color='green', linewidth=2.5)
plt.title('Average Global GDP per Capita Trend', fontsize=14)
plt.ylabel('Average GDP per Capita (USD)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['Unemployment Rate (%)'], color='orange', width=0.5)
plt.title('Distribution of Global Unemployment Rate (%)', fontsize=14)
plt.ylabel('Unemployment Rate (%)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Unemployment Rate (%)', y='Inflation (CPI %)', alpha=0.5)
plt.title('Inflation (CPI %) vs. Unemployment Rate (%)', fontsize=14)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.ylim(-5, 25) # Zooming in to exclude extreme hyperinflation outliers
plt.show()
```
